````python
sqlCreateTable = 'CREATE TABLE tbl ();'
p = re.compile(' +|-') # blank
for i, v in enumerate(df.columns):
  if df.iloc[1:2,i].dtypes == 'object':
    #print(f'{i}, {v}: object, max: {(int(df[v].map(lambda x: len(x)).max()/10) + (df[v].map(lambda x: len(x)).max()%10>0))*10}') # roundup: int(21 / 5) + (21 % 5 > 0)
    if df[v].map(lambda x: len(x)).max() >= 100:
      sqlCreateTable = sqlCreateTable[:-2] + '\n' + p.sub('_', v) + ' text,' + sqlCreateTable[-2:]
      #print('\ttext is recommended')
    else:
      sqlCreateTable = sqlCreateTable[:-2] + '\n' + p.sub('_', v) + ' varchar(' + str((int(df[v].map(lambda x: len(x)).max()/10) + (df[v].map(lambda x: len(x)).max()%10>0))*10) + '),' + sqlCreateTable[-2:]
  elif df.iloc[1:2,i].dtypes == 'int64':
    #print(f'{i}, {v}: int64')
    sqlCreateTable = sqlCreateTable[:-2] + '\n' + p.sub('_', v) + ' integer,' + sqlCreateTable[-2:]
  elif df.iloc[1:2,i].dtypes == 'float64':
    #print(f'{i}, {v}: float64')
    sqlCreateTable = sqlCreateTable[:-2] + '\n' + p.sub('_', v) + ' real,' + sqlCreateTable[-2:]
  elif df.iloc[1:2,i].dtypes == 'datetime64[ns]':
    #print(f'{i}, {v}: datetime64[ns]')
    sqlCreateTable = sqlCreateTable[:-2] + '\n' + p.sub('_', v) + ' timestamp,' + sqlCreateTable[-2:]
  else:
    print('...')
sqlCreateTable = sqlCreateTable[:-3] + sqlCreateTable[-2:]
print(sqlCreateTable)
````

````python
from sqlalchemy import MetaData
from sqlalchemy_schemadisplay import create_schema_graph
# create the pydot graph object by autoloading all tables via a bound metadata object
graph = create_schema_graph(
    metadata=MetaData('postgres://yhr:yhr@localhost/postgres'),
    show_datatypes=False, # The image would get nasty big if we'd show the datatypes
    show_indexes=False, # ditto for indexes
    rankdir='LR', # From left to right (instead of top to bottom)
    concentrate=False # Don't try to join the relation lines together
)
graph.write_png('dbschema.png') # write out the file
````

````python
from sqlalchemy import create_engine, MetaData
from sqlalchemy_schemadisplay import create_schema_graph

# Create an engine and reflect the database
#engine = create_engine('postgresql://my_user:mypassword@localhost:5432/my_db')
engine = create_engine('postgresql://yhr:yhr@localhost:5432/postgres')
metadata = MetaData(schema='yhr')
metadata.reflect(bind=engine)

# Now create the schema graph
graph = create_schema_graph(
    metadata=metadata,
    show_datatypes=True,
    show_indexes=True,
    rankdir='LR',
    concentrate=False,
    #show_schema_name=True
)
graph.write_png('dbschema.png')
````
