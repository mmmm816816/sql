## smart_customer

```python
c = conn.cursor()
c.execute("""
          create table smart_customer (
            customer_name varchar(20),
            customer_id integer,
            mail text,
            gas varchar(20),
            agency varchar(20),
            category varchar(20)
          );
          """)
conn.commit() # create/insert into/update
# c.cursor()
```

```python
df = pd.read_csv('smart_customer_utf-8.csv') # encoding = "SHIFT-JIS"
```

```python
df
```

```python
df.columns
```

```python
c = conn.cursor()
for index, row in df.iterrows():
     c.execute(f"""insert into smart_customer (customer_name, customer_id, mail, gas, agency, category) values (
          '{row.customer_name}', 
          {row.customer_id}, 
          '{row.mail}',
          '{row.gas}',
          '{row.agency}',
          '{row.category}')"""
     )
conn.commit()
```

```python
c = conn.cursor()
c.execute("""
SELECT
  column_name, data_type -- *
FROM
  information_schema.columns
WHERE
  table_schema = 'yhr' and table_name = 'smart_customer';
          """)
query_result = c.fetchall()
print(query_result)
c.close()
```

## info

```python
c = conn.cursor()
c.execute("""
          create table info (
            mail text
          );
          """)
conn.commit() # create/insert into/update
```

```python
df = pd.read_csv('info_utf-8.csv') # encoding = "SHIFT-JIS"
df
```

```python
c = conn.cursor()
for index, row in df.iterrows():
     c.execute(f"""insert into info (mail) values (
          '{row.mail}')"""
     )
conn.commit()
```

```python
c = conn.cursor()
c.execute("""
with have_mail_info as (
  select
    mail,
    case when i.mail is null then 'mail delivery' else 'e-mail' end
  from
    smart_customer c
    left outer join
    info i
    on c.mail = i.mail
), case1 as (
  select
    'case1' as notification
    have_mail_info.mail
  from
    have_mail_info
  where
    gas = 'True'
), case2 as (
  select
    'case2' as notification
    have_mail_info.mail
  from
    have_mail_info
  where
    agency = '0.0' and category = 'keizoku'
), case3 as (
  select
    'case3' as notification
    have_mail_info.mail
  from
    have_mail_info
  where
    agency = '1.0' and category = 'keizoku'
), case1_distinct as (
  select distinct on (mail)
    *
  from
    case1
)
select
  *
from
  case1_distinct
union all
select
  *
from
  case2
union all
select
  *
from
  case3
          """)
query_result = c.fetchall()
print(query_result)
c.close()
```
